```python

from google.colab import files
uploaded = files.upload()

import pandas as pd


file_paths = {
    "Antofagasta": "familia antofagasta.csv",
    "Magallanes": "familia magallanes.csv",
    "Metropolitana": "familia metropolitana.csv",
    "Valparaíso": "familia valparaiso.csv"
}


def procesar_archivo(path, region):
    df = pd.read_csv(path)
    registros = []
    for _, row in df.iloc[1:].iterrows():
        partes = row.iloc[0].split(";")
        tipo = partes[0]
        valores = partes[1:]
        for i, año in enumerate(range(2005, 2005 + len(valores))):
            if año == 2025:
                continue
            try:
                casos = int(valores[i].replace('.', '').strip())
            except:
                casos = None
            registros.append({
                "Región": region,
                "Tipo de Delito": tipo,
                "Año": año,
                "Casos": casos
            })
    return pd.DataFrame(registros)

dfs = {region: procesar_archivo(path, region) for region, path in file_paths.items()}


!pip install plotly --quiet
import plotly.express as px

for region, df_region in dfs.items():
    fig = px.line(
        df_region,
        x="Año",
        y="Casos",
        color="Tipo de Delito",
        title=f"Evolución de delitos en {region} (2005–2024)",
        markers=True,
        line_shape="spline"
    )
    fig.update_layout(
        template="plotly_white",
        legend_title_text="Tipo de Delito",
        hovermode="x unified",
        xaxis=dict(tickmode='linear', dtick=1)
    )
    fig.show()
```
